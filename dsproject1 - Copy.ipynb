```python
import pandas as pd
import numpy as np

# Load the dataset
mastercopy = pd.read_csv(r'C:\Users\HP\OneDrive\Documents\data_science_project\mastercopy_with_nans.csv')

# Create a copy named 'mean'
mean_copy = mastercopy.copy()

# Loop through all columns and replace NaNs with the mean of the column
for column in mean_copy.columns:
    if mean_copy[column].dtype in ['float64', 'int64']:  # Check if the column is numeric
        if mean_copy[column].isna().sum() > 0:  # Check if the column contains NaN values
            mean_val = mean_copy[column].mean()  # Calculate the mean of the column
            mean_copy[column] = mean_copy[column].fillna(mean_val)  # Replace NaNs with the mean

# Display the columns after replacing NaNs
print("Data after replacing NaNs with mean in 'mean' copy:")
print(mean_copy.head(15))  # Display the first 15 rows to check the changes

# Save the 'mean' copy DataFrame to an Excel file
output_path_mean = r'C:\Users\HP\OneDrive\Documents\data_science_project\mean_copy_with_nans_replaced.xlsx'
mean_copy.to_excel(output_path_mean, index=False)

# Save the 'mean' copy DataFrame to a CSV file
output_path_mean_csv = r'C:\Users\HP\OneDrive\Documents\data_science_project\mean_copy_with_nans_replaced.csv'
mean_copy.to_csv(output_path_mean_csv, index=False)

print(f"\nThe updated 'mean' dataset has been saved to: {output_path_mean}")
print(f"\nThe updated 'mean' dataset has been saved to: {output_path_mean_csv}")
```

```python
#Standardized Values 
import pandas as pd
import numpy as np

# Load the dataset
mastercopy = pd.read_csv(r'C:\Users\HP\OneDrive\Documents\data_science_project\mastercopy_with_nans.csv')

# Create a copy for standardization
standardized_copy = mastercopy.copy()

# Replace NaN values with standardized values (Z-score)
for column in standardized_copy.columns:
    if standardized_copy[column].dtype in ['float64', 'int64']:  # Check if the column is numeric
        if standardized_copy[column].isna().sum() > 0:  # Check if the column contains NaN values
            mean_val = standardized_copy[column].mean()  # Mean of the column
            std_val = standardized_copy[column].std()  # Standard deviation of the column
            # Standardize by replacing NaNs with (value - mean) / std
            standardized_copy[column] = standardized_copy[column].fillna((mean_val - mean_val) / std_val)

# Save the 'standardized' copy DataFrame to an Excel file
output_path_standardized = r'C:\Users\HP\OneDrive\Documents\data_science_project\standardized_copy_with_nans_replaced.xlsx'
standardized_copy.to_excel(output_path_standardized, index=False)

print(f"\nThe 'standardized' copy dataset has been saved to: {output_path_standardized}")
```

```python
import pandas as pd
import numpy as np
import time
from sklearn.model_selection import train_test_split, cross_validate
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score

# Load the dataset
mean_copy = pd.read_excel(r'C:\Users\HP\OneDrive\Documents\data_science_project\mean_copy_with_nans_replaced.xlsx')
csv_file='output.csv'
mean_copy.to_csv(csv_file,index=False)

# Create a copy of the dataset for processing
data =mean_copy.copy()

# Define the target column (adjust this to your target column)
target_column = 'apparent_temperature'  # Change this to your actual target column name
X = data.drop(columns=[target_column])  # Features (X)
y = data[target_column]  # Target (y)

# Step 1: Handle Date Columns
date_columns = X.select_dtypes(include=['object', 'datetime']).columns  # Find date-like columns

# Convert date columns into numeric features
for col in date_columns:
    # Convert string dates to datetime if not already in datetime format
    X[col] = pd.to_datetime(X[col], errors='coerce')
    
    # Extract features from the date columns like year, month, day, day of the week, etc.
    X[col + '_year'] = X[col].dt.year
    X[col + '_month'] = X[col].dt.month
    X[col + '_day'] = X[col].dt.day
    X[col + '_dayofweek'] = X[col].dt.dayofweek  # Monday=0, Sunday=6
    
    # Drop the original date column
    X = X.drop(columns=[col])

# Step 2: Handle Categorical Columns (Label Encoding)
for col in X.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col].astype(str))
    # Drop rows with NaN values
X_clean = X.dropna()
y_clean = y[X_clean.index]  # Make sure target variable aligns

# Train the model
model = LogisticRegression()
model.fit(X_clean, y_clean)


# Step 3: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=50)

# Step 4: Define Models
models = {
    "Logistic Regression": LogisticRegression(max_iter=2000),
    "Support Vector Machine": SVC(probability=True),
    "Decision Tree Classifier": DecisionTreeClassifier(),
    "Naïve Bayes Classifier": GaussianNB(),
    "K Nearest Neighbour Classifier": KNeighborsClassifier(),
    "Random Forest Classifier": RandomForestClassifier()
}

# Step 5: Train and Evaluate Models using Cross-Validation
results = []

for name, model in models.items():
    start_time = time.time()  # Start the timer using time.time()
    
    # Perform cross-validation
    scores = cross_validate(model, X_train, y_train, cv=5, scoring='accuracy', return_train_score=False)
    
    end_time = time.time()  # End the timer
    
    # Compute the average of the accuracy across the folds
    accuracy = np.mean(scores['test_score'])
    elapsed_time = end_time - start_time
    
    # Store the results
    results.append({
        'Model': name,
        'Accuracy': accuracy,
        'Time (s)': elapsed_time
    })
    
    print(f"{name} completed in {elapsed_time:.4f} seconds with Accuracy: {accuracy:.4f}")

# Step 6: Convert Results to DataFrame and Display
results_df = pd.DataFrame(results)

# Display the results
print("\nModel Evaluation Results:")
print(results_df)

# Optionally, save the results to a CSV file
output_path = r'C:\Users\HP\OneDrive\Documents\data_science_project\model_results.csv'
results_df.to_csv(output_path, index=False)

print(f"\nThe model evaluation results have been saved to: {output_path}")
```
