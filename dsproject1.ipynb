```python
#mean
import pandas as pd
import numpy as np

# Load the dataset
weather = pd.read_csv(r'C:\Users\HP\OneDrive\Documents\data science project\hourly_weather_data.csv')

# Create a copy named 'mean'
mean_copy = weather.copy()

# Loop through all columns and replace NaNs with the mean of the column
for column in mean_copy.columns:
    if mean_copy[column].dtype in ['float64', 'int64']:  # Check if the column is numeric
        if mean_copy[column].isna().sum() > 0:  # Check if the column contains NaN values
            mean_val = mean_copy[column].mean()  # Calculate the mean of the column
            mean_copy[column] = mean_copy[column].fillna(mean_val)  # Replace NaNs with the mean

# Display the columns after replacing NaNs
print("Data after replacing NaNs with mean in 'mean' copy:")
print(mean_copy.head(15))  # Display the first 15 rows to check the changes

# Save the 'mean' copy DataFrame to an Excel file
output_path_mean = r'C:\Users\HP\OneDrive\Documents\data science project\mean_copy_with_nans_replaced.xlsx'
mean_copy.to_excel(output_path_mean, index=False)

print(f"\nThe updated 'mean' dataset has been saved to: {output_path_mean}")
```

```python
#Standard Deviation
import pandas as pd
import numpy as np

# Load the dataset
weather = pd.read_csv(r'C:\Users\HP\OneDrive\Documents\data science project\hourly_weather_data.csv')

# Create a copy for std (standard deviation) replacement
std_copy = weather.copy()

# Replace NaN values with the standard deviation of each column
for column in std_copy.columns:
    if std_copy[column].dtype in ['float64', 'int64']:  # Check if the column is numeric
        if std_copy[column].isna().sum() > 0:  # Check if the column contains NaN values
            std_val = std_copy[column].std()  # Calculate the standard deviation of the column
            std_copy[column] = std_copy[column].fillna(std_val)  # Replace NaNs with the standard deviation

# Save the 'std' copy DataFrame to an Excel file
output_path_std = r'C:\Users\HP\OneDrive\Documents\data science project\std_copy_with_nans_replaced.xlsx'
std_copy.to_excel(output_path_std, index=False)

print(f"\nThe 'std' copy dataset has been saved to: {output_path_std}")
```

```python
#Standardized Values 
import pandas as pd
import numpy as np

# Load the dataset
weather = pd.read_csv(r'C:\Users\HP\OneDrive\Documents\data science project\hourly_weather_data.csv')

# Create a copy for standardization
standardized_copy = weather.copy()

# Replace NaN values with standardized values (Z-score)
for column in standardized_copy.columns:
    if standardized_copy[column].dtype in ['float64', 'int64']:  # Check if the column is numeric
        if standardized_copy[column].isna().sum() > 0:  # Check if the column contains NaN values
            mean_val = standardized_copy[column].mean()  # Mean of the column
            std_val = standardized_copy[column].std()  # Standard deviation of the column
            # Standardize by replacing NaNs with (value - mean) / std
            standardized_copy[column] = standardized_copy[column].fillna((mean_val - mean_val) / std_val)

# Save the 'standardized' copy DataFrame to an Excel file
output_path_standardized = r'C:\Users\HP\OneDrive\Documents\data science project\standardized_copy_with_nans_replaced.xlsx'
standardized_copy.to_excel(output_path_standardized, index=False)

print(f"\nThe 'standardized' copy dataset has been saved to: {output_path_standardized}")
```

## Deduplication of NASA

Much of the samples that exist in the dataset are from the same date and time. These samples just have slightly different `lat/long`, and their corresponding `brightness`.<br />
The rest of the features like `scan`, `track`, `satellite` etc. are the same. These samples does not add variance/diversity to the dataset, and may in turn lead to bias or create artificial importance of these fixed features.

Therefore, its better to remove such samples.
